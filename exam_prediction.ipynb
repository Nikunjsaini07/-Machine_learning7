```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt  
```

```python
df = pd.read_csv("student.csv")
```

```python
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})
 
```

```python
df.head()
```

```python
df = df.drop(columns=['student_id'])
```

```python
df.head()
```

```python
df['part_time_job'] = df['part_time_job'].map({'No': 0, 'Yes': 1})
```

```python
df['diet_quality'].unique()
```

```python
diet_map = {
    'Poor': 0,
    'Fair': 1,
    'Good': 2,
    
}
df['diet_quality'] = df['diet_quality'].map(diet_map)
```

```python

df.head()
```

```python
df['internet_quality'].unique() 
```

```python
internet_map = {
    'Poor': 0,
    'Average': 1,
    'Good': 2,
    
}
df['internet_quality'] = df['internet_quality'].map(internet_map)
```

```python
df.head()
```

```python
df['extracurricular_participation'] = df['extracurricular_participation'].map({'Yes': 1, 'No': 0})
```

```python
df['parental_education_level'].unique()
```

```python
parent_map = {
    'nan': 0,
    'High School': 1,
    'Bachelor': 2,
    'Master' : 3 
    
}
df['parental_education_level'] = df['parental_education_level'].map(parent_map)
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'], errors='ignore')
```

```python
df.head(4)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['parental_education_level'] = df['parental_education_level'].fillna(1)
```

```python
df['gender'] = df['gender'].fillna(0)
```

```python
df = df.astype("int64")  
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

```python
df.corr()
```

```python
df.drop(columns=['age' , 'gender', 'part_time_job' , 'diet_quality', 'internet_quality','parental_education_level', 'internet_quality',  'extracurricular_participation'], inplace=True)
```

```python
df.isnull().sum()
```

```python
x = df.iloc[: , : -1 ]
```

```python
y = df.iloc[: , -1]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x, y,                
    test_size=0.2,       
    random_state=42,    
              
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_pred= model.predict(X_test_scaled)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)

```

```python
r2
```

```python
from sklearn.metrics import mean_squared_error


rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
```

```python
plt.scatter(y_test, y_pred, color='blue', alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='black', linestyle='--')
plt.xlabel('Actual Exam Score')
plt.ylabel('Predicted Exam Score')
plt.title('Actual vs Predicted Exam Scores')
plt.grid(True)
plt.show()
```
